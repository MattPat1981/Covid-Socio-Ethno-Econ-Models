## HelloModel
As the 'hello_world' of machine learning, the linear regression model (also known as Ordinary Least Squares) is a linear model that allows us to look at many different features and predict a single continuous variable. In this case, we are trying to use the linear regression model to predict how many deaths a county will see as a result of covid-19.

## Features
In order to create our predictions, this notebook will demonstrate how we selected each feature and will mention a few of the features that we chose not to use in our predictions. Each feature will hopefully help us to see linearity in our model when plotted on an xy axis. There should be independence between features (ie no feature should be an aggregate of another, such as 'deaths per day' and 'deaths cumulative'). Also, we'll want to see normality in the features when we look at a histogram. While normality won't be perfect, we'll see that nothing really is in predictive modeling. Finally, we really would like to see homoscedasticity as a result of the model, or a minimum of outliers in the data. This way we are modeling the behavior and not the noise.

## The Noise
It should be noted before we dive too deep in to this that places like New York City and counties in Northern New Jersey are expected to appear as outliers in this model. This can either skew the results, or we can use the prior knowledge about these areas to our advantage in future predictions. 

Alternatively, we can take intutive information about these counites like population density, the reliance on trains and busses for public transportation, and the abundance of 'essential' workers whose jobs require them to work in close quarters with other people; and compare these to the traits of rural counties in Georgia, Louisianna, Florida, and New Mexico where we also see an abundance of deaths due to Covid-19 but do not see many of the aforementioned traits. What else do these areas have in common?

The noise in our model may tell a story or it might try to lead us astray. Let's begin.

OK, so we have actually gotten worse at predicting new information using the additional data, and have successfully overfit our model. We got a training R2 score of 79.47% and a testing score of 66.57%. There is too much variance in our model. We can now package up a modeling function and then tweak the inputs.

```python
def model_this(df, features, target, test_size):
    '''
    returns a linear regression model instance and prints the R2 score for training and testing
    df        is a pandas dataframe from which to build the model
    feautres  is a list used to create your X matrix
    target    is a string, used to create the y list, the target column to predict
    '''
    
    # if re-packaging this funciton, you'll need to import your libraries here
    # if you're copying this function, give credit to Matt Paterson, 
    # hello@hireMattPaterson.com, HireMattPaterson.com
    
    X = df[features]
    y = df[target]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                       test_size = test_size,
                                                       random_state=603)
    model = LinearRegression()
    model.fit(X_train, y_train)
    
    train_score = model.score(X_train, y_train)
    test_score = model.score(X_test, y_test)

    print("Training Score : ", train_score)
    print("Testing Score  : ", test_score)
    
    return model
```

## Root Mean Squared Error
The RMSE, or Root Mean Squared Error, is a commonly used metric in Linear Regression. In short, the Root Mean Squared Error is used to determine how far from the acutal number we allow our predictions to be. So in this case, an RMSE of 40 would mean that we predicted the number of Covid cases (or deaths) in a county within 40 of its actual number.

```python
def model_and_pred(df, features, target, test_size):
    '''
    returns a linear regression model instance and prints the R2 score for training and testing
    df        is a pandas dataframe from which to build the model
    feautres  is a list used to create your X matrix
    target    is a string, used to create the y list, the target column to predict
    '''
    
    # if re-packaging this funciton, you'll need to import your libraries here
    # if you're copying this function, give credit to Matt Paterson, 
    # hello@hireMattPaterson.com, HireMattPaterson.com
    
    X = df[features]
    y = df[target]
    
    X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                       test_size = test_size,
                                                       random_state=603)
    model = LinearRegression()
    model.fit(X_train, y_train)
        
    y_predictions = model.predict(X_test)
    residuals = y_test - y_predictions

    #reply = input('Press any key to see the error metrics...')

    MSE = mean_squared_error(y_test, y_predictions)

    print('r2 score :               ', model.score(X_test, y_test))
    print('Mean Squared Error:      ', MSE)
    print('Root Mean Squared Error: ', MSE**(1/2))
    print('cross_val_score:  ', cross_val_score(model, X_test, y_test).mean())
    residuals.hist()
    plt.title('Histogram of residuals (Predictions versus Actuals)')
    #plt.legend()

    
    return model
```

```python
#df.columns
```

```python
targets = ['deaths', 'confirmed_cases']
lineup = "*" * 30
order = ['first', 'second', 'third']
# target = 'confirmed_cases'

for target in targets:
    print(f'\n{lineup}\nWhen predicting the number of {target} in a county:')

    for entry in entries:
        print(f'\nUsing {entry};\n with {big_test_size:.0%}:')
        model_a = model_and_pred(df, entry, target, big_test_size)

        print(f'\nUsing {entry};\n with {small_test_size:.0%}:')
        model_b = model_and_pred(df, entry, target, small_test_size)
        print(lineup)
```
